#### 12. Join users to transactions, keeping all rows from transactions and only matching rows from users (left join) 

    Expected Output:


#### Solution

```python
pd.merge(transactions, users, how='left',on='UserID').head(10)
```

#### 13. Which transactions have a UserID not in users?
    Expected Output:


#### Solution

```python
transacuser= transactions.join(users,on='UserID',how='left',lsuffix='_transac',rsuffix='_user')
```

```python
transacuser.loc[transacuser['UserID_user'].isnull(),]
```

#### 14. Join users to transactions, keeping only rows from transactions andusers that match via UserID (inner join) 
    Expected Output:
    

#### Solution

```python
pd.merge(transactions,users,on='UserID',how='inner').head(7)
```

#### 15. Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
    Expected Output:
    

#### Solution

```python
pd.merge(transactions, users, how='outer',on='UserID').head(12)
```

#### 16. Determine which sessions occurred on the same day each user registered 
    Expected Output:
    

#### Solution

```python
pd.merge(users,sessions,how='inner',left_on=['UserID','Registered'],right_on=['UserID','SessionDate']).head(2)
```

####  17. Build a dataset with every possible (UserID, ProductID) pair (cross join) 
    ExpectedOutput:


#### Solution

```python
users['key']=1
products['key']=1
usersproducts = pd.merge(users,products,how='outer',on='key')
usersproducts.loc[:,['UserID','ProductID']].head(13)
```

#### 18. Determine how much quantity of each product was purchased by each user




#### Solution

```python
usersproductstransac = pd.merge(usersproducts,transactions,how='outer',on=['UserID','ProductID'])
usersproductstransac.loc[:,['UserID','ProductID','Quantity']].fillna(0.0).head(15)
```

#### 19. For each user, get each possible pair of pair transactions (TransactionID1, TransacationID2)
    Expected Output:


#### Solution

```python
pd.merge(transactions,transactions,how = 'outer',on = 'UserID').head(15)
```
