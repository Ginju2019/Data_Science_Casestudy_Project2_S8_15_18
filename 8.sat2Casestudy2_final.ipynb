__Problem Statement 18:__<br><br>
In a certain town, there are about one million eligible voters. A simple random sample of 10,000 eligible voters was chosen to study the relationship between gender and participation in the last election. The results are summarized in the following 2X2 (read two by two) contingency table:











We would want to check whether being a man or a woman (columns) is independent of having voted in the last election (rows). In other words, is “gender and voting independent”?



__Solution:__

```python
#applying chi square test here
```

```python
# Defining Null & Alternative Hypothesis
# H0: Votes are independent of the gender
# H1: Votes are dependent of the gender
```

```python
#finding total votes and total men/women
Total_men=2792+1486 # total no of men
Total_women=3591+2131 # total no of women
Total_vote = 2792+ 3591 #total votes by men and women
Total_notvote = 1486+2131 # total no votes by men and women
Grand_Total= 10000 # sample size
```

```python
# finding the Expected value of the genders voted by (row total * column total)/ Grand total
Expected_men_votes=(Total_vote*Total_men)/Grand_Total
Expected_women_votes=(Total_vote*Total_women)/Grand_Total
```

```python
# finding the Expected value of the genders notvoted by (row total * column total)/ Grand total
Expected_men_not_votes=(Total_notvote*Total_men)/Grand_Total
Expected_women_not_votes=(Total_notvote*Total_women)/Grand_Total
```

```python
print("Expected_men_votes: ", Expected_men_votes,
      "Expected_women_votes: ",Expected_women_votes , 
      "Expected_men_not_votes: ", Expected_men_not_votes, 
      "Expected_women_not_votes:", Expected_women_not_votes)
```

```python
ChiSqr = (2792-Expected_men_votes)**2/Expected_men_votes + (3591 - Expected_women_votes)**2/Expected_women_votes + (1486 - Expected_men_not_votes)**2/Expected_men_not_votes + (2131 - Expected_women_not_votes)**2/Expected_women_not_votes
print("ChiSqr: ", ChiSqr)
```

#### Reject Null Hypothesis as the chisquare value is higher than 5%, hence their is  high probability of voting dependent on the gender
